```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

from data_process import *
```

```python
filepath = '../spectra_files/psg_trn.txt'
data = psg_to_numpy(filepath)
print(data)
```

```python
plt.plot(data[:, 0], data[:,1])
plt.xlabel('Wavelength')
plt.ylabel('Fractional transmittence')
```

```python
subset = data[9000:9500]
```

```python
peaks, properties = find_peaks(-subset[:, 1], height=-0.975)

plt.plot(subset[:, 0], subset[:, 1], 'o-', markersize=2)
plt.scatter(subset[peaks, 0], subset[peaks, 1], color='tab:orange')
plt.xlabel('Wavelength')
plt.ylabel('Fractional transmittence')
```

```python
plt.plot(subset[:, 0], -subset[:, 1], 'o-', markersize=2)
plt.xlabel('Wavelength')
plt.ylabel('Fractional transmittence')
```
