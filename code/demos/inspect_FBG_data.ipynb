# Inspect FBG Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_path = 'downloadable_data'
files = {1: ('C2H2', 'C03183_0_T'),
         2: ('C2H2', 'C03191_0_T'),
         3: ('C2H2', 'C03192_0_T'),
         4: ('C2H2', 'C03193_0_T'),
         5: ('HCN', 'C03072_0_RT'),
         6: ('HCN', 'C03111_0_RT'),
         7: ('HCN', 'C03112_0_RT'),
         8: ('HCN', 'C03113_0_RT'),
         9: ('HCN', 'C03114_0_RT')}
```

```python
file = 1
data = pd.read_csv('{}/{}/{}.csv'.format(data_path, files[file][0], files[file][1]))
```

## Conversions between dB and fractional transmittence

* Using the power ratio conversion: https://en.wikipedia.org/wiki/Decibel

```python
# def dB_to_frac(x):
#     return 10**(x/10)

data['Tx'] = 10**(data['Tx (dB)']/10)
data['Rx'] = 10**(data['Rx (dB)']/10)
```

## Transmittance

```python
plt.figure(figsize=(10, 3))
plt.subplot(1, 2, 1)
plt.plot(data['Wavelength (nm)'], data['Tx (dB)'])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmittance (dB)')

plt.subplot(1, 2, 2)
plt.plot(data['Wavelength (nm)'], data['Tx'])
plt.ylim([0, 1])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Fractional Transmittance')
plt.show()
```

## Reflectance

```python
plt.figure(figsize=(10, 3))
plt.subplot(1, 2, 1)
plt.plot(data['Wavelength (nm)'], data['Rx (dB)'])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (dB)')

plt.subplot(1, 2, 2)
plt.plot(data['Wavelength (nm)'], data['Rx'])
plt.ylim([0, 1])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Fractional Reflectance')
plt.show()
```

## Transmittance + Reflectance

We expect this to be at 1

```python
plt.figure(figsize=(3, 3))
plt.plot(data['Wavelength (nm)'], data['Tx'] + data['Rx'])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmittance + Reflectance')
plt.show()
```
